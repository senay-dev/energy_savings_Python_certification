<p>The government would like to motivate citizens by sharing the total energy that the combined recycling efforts have saved every year. They have asked you to help them.</p>

```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
stats03_17 = pd.read_csv('datasets/wastestats.csv')
stats18_19 = pd.read_csv('datasets/2018_2019_waste.csv')
energy_saved = pd.read_csv('datasets/energy_saved.csv')
```

```python
# ['Plastic','Glass','Ferrous Metal','non-ferrous metal']
# 03_17 Plastic,Glass,Ferrous Metal, Non-ferrous Metal
# 18_19 Plastics, Glass, Ferrous Metal, Non-ferrous Metal
stats03_17.head()
```

```python
names= ['Plastic','Glass','Ferrous Metal','Non-ferrous metal','Non-ferrous metal','Plastics','Ferrous metal','Non-ferrous Metals','Ferrous Metals','Non-ferrous metals']
stats1 = stats03_17[stats03_17.waste_type.isin(names)].reset_index()
stats2 = stats18_19[stats18_19['Waste Type'].isin (['Plastics','Glass','Ferrous Metal','Non-Ferrous Metal'])].reset_index()                
               
```

```python
energy_cleaned = pd.DataFrame({'Material':energy_saved.iloc[2,1:].reset_index().iloc[:,1],'kWh':energy_saved.iloc[3,1:].reset_index().iloc[:,1]})
names= ['Non-ferrous metal','Plastics','Ferrous metal','Non-ferrous Metals','Ferrous Metals','Non-ferrous metals']
vals = ['14000 Kwh','5774 Kwh','642 Kwh','14000 Kwh','642 Kwh','14000 Kwh']
dic = {'Material':names,'kWh':vals}
energy_cleaned = energy_cleaned.append(pd.DataFrame(dic,index=[0,1,2,3,4,5]),ignore_index=True)
energy_cleaned.set_index('Material',inplace=True)
energy_cleaned['kWh']=energy_cleaned['kWh'].str.slice(start=0,stop=-4).astype('int64')
stats1_list = []
for i in stats1.waste_type:
    stats1_list.append(energy_cleaned.loc[i].iloc[0])

stats2_list = []
for i in stats2['Waste Type']:
    stats2_list.append(energy_cleaned.loc[i].iloc[0])
```

```python
stats1['energy_per_type'] = pd.DataFrame({'energy_per_type':stats1_list})
stats2['energy_per_type'] = pd.DataFrame({'energy_per_type':stats2_list})
```

```python
stats1['total_energy_saved'] = stats1['total_waste_recycled_tonne']*stats1['energy_per_type']
stats2['total_energy_saved'] = stats2["Total Recycled ('000 tonnes)"]*stats2['energy_per_type']*1000
```

```python
total_waste1 = stats1.groupby(['year']).agg({'total_energy_saved':['sum']})
total_waste2 = stats2.groupby(['Year']).agg({"total_energy_saved":['sum']})
```

```python
annual_energy_savings = pd.concat([total_waste1.loc[2015:],total_waste2])
annual_energy_savings = annual_energy_savings.reset_index()
annual_energy_savings.columns = ['year','total_energy_saved']
annual_energy_savings = annual_energy_savings.set_index('year')
annual_energy_savings
```
